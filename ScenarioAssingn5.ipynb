```python
def largest_prime_factor(n):
    largest = 1
    i = 2
    while i * i <=n:
        while n % i == 0 :
            largest = 1
            n //= i
        i +=1 
    if n> 1:
        largest = n
    return largest
num = int(input("Enter number:"))
print("largest_prime_factor:", largest_prime_factor(num))
            
        
```

```python
def roman_to_int(s):
    roman = {"S":20,"M":25,"XL":35,
                 "XL":45,"XXL":55}
total = 0
for i in range(len(s)):
     if i + 1 < len(s) and roman [s[i]]< roman [s[i + 1]]:
        total -= roman [s[i]]
     else:
        total += roman [s[i]]
return total
roman_num = input("Enter Roman numeral:")
print("Integer Num:",roman_to_int(roman_num))
```

```python
def longest_common_prefix(words):
    prefix = words[0]

    for word in words[1:]:
        while not word.startswith(prefix):
            prefix = prefix[::-1]
            if prefix == "":
                return ""
    return prefix

words = input("Enter Words:").split()
print("Longest Common Prefix:",longest_common_prefix(words))
```
