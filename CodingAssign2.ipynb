```python
def reversed_Name(name):
    return name [::-1]
print(reversed_Name("Hello"))
   
```

```python
def fibonacci(n):
    sequence =[0,1]
    for i in range (2,n):
        sequence.append(sequence[-1]+ sequence[-2])
    return sequence
n_terms = 10
fib_sequence = fibonacci(n_terms)
print(*fib_sequence)
```

```python
def is_palindrome(s):
    return s == s[::-1]
word = "madam"
if is_palindrome(word):
    print(f"{word} is a palindrome")
else:
    print(f"{word} is not a palindrome")
```

```python
string = input("Enter a String:")
vowels = "aeiouAEIOU" 
count = 0
for char in string:
    if char in vowels:
        count += 1 
print("Number of Vowels:",count)
```

```python
my_list = [1,2,2,3,4,4,5]
unique_list = []
for item in my_list:
    if item not in unique_list:
        unique_list.append(item)

print(unique_list)
```
