```python
def percentage():

    m1=int(input("Subject1= "))
    m2=int(input("Subject2= "))
    m3=int(input("Subject3= "))
    m4=int(input("Subject4= "))
    m5=int(input("Subject5= "))
    
    Total = m1+m2+m3+m4+m5
    print("Total : ",m1+m2+m3+m4+m5)
    Percent = (Total / 500) * 100
    print("Percentage : ",Percent)

percentage()    
```
