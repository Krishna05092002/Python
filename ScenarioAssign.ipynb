```python
def Retail_store():
    store = int(input("Enter the total Order:"))
    if store > 100 :
        discount = store * 0.20   
    else:
        discount = 0

    final_amount = store - discount
    return final_amount


Retail_store()
```

```python
def withdraw(balance):
    amount = float(input("Enter amount to withdraw: "))
    if amount <= balance:
        balance -= amount
        print(" withdrawn successfully.")
        print("Updated balance:", balance)
    else:
        print("Insufficient funds.")
    return balance

balance = 10000
balance = withdraw(balance)
```
